```python
colnames = ['Chromosome', 'Start', 'End', 'Name', '20nt hits', 'Align', 'Chromosome targets', 'Sequence', 'Tm']
path = "/FISH/CTRL_L/CTRL_L_hg38_probes.csv"
df = pd.read_csv(path, names=colnames, sep=';', header=None)
```

```python
overhang = "GCTATCGTTCGTTCGAGGCC"
df["Sequence with Overhang"] = overhang + df["Sequence"]
df
```

```python
df.to_csv("Altius_CTRL.csv")
```

```python
#load data CTRL_L PaintSHOP
path2 = "/FISH/CTRL_L/CTRL_L_PaintSHOP-full_10KB.txt"
colnames2 = ["Chromosome", "Start", "End", "Sequence", "Tm", "On Target", "Off Target", "Repeat Seq", 
             "Prob", "Max Kmer", "Probe Strand", "Target"]
df_2 = pd.read_table(path2, names = colnames2, sep='\t')
```

```python
df_2["Sequence with Overhang"] = overhang + df_2["Sequence"]
df_2
```

```python
start_min = df['Start'].min()
end_max = df['End'].max()
offset = 5500
df_2_filt = df_2[(df_2["Start"]>=start_min-offset) & (df_2["End"]<=end_max+offset)]

df_2_filt
```

```python
df_2_filt["Length"] = df_2_filt['Sequence'].apply(len)
df["Length"] = df['Sequence'].apply(len)
```

```python
df_2_central = df_2_filt[(df_2_filt["Start"]>=start_min) & (df_2_filt["End"]<=end_max)]

df_2_offset = df_2_filt[~((df_2_filt["Start"]>=start_min) & (df_2_filt["End"]<=end_max))]
```

```python
df_2_central.to_csv("PaintShop_CTRL_central.csv")
df_2_offset.to_csv("PaintShop_CTRL_offset.csv")
```

```python
from Bio.SeqUtils import GC
from Bio.SeqUtils import MeltingTemp as mt

def corrected_tm(seq):
    tm = mt.Tm_NN(seq, Na=390)
    tm_corr = mt.chem_correction(tm, fmd=50)
    return tm_corr
```

```python
df["Tm"] = df['Sequence'].apply(corrected_tm)
df["GC"] = df['Sequence'].apply(GC)
df_2_filt["GC"] = df_2_filt['Sequence'].apply(GC)
```

```python
df['source']='Altius'
df_2_filt['source']='PaintSHOP'
```

```python
df_stack=pd.concat([df, df_2_filt], axis=0).reset_index(drop=True)
df_stack.to_csv("Altius_Paintshop.csv")
```

```python
plt.figure(figsize = (15,6))
sns.set(font_scale=1.5)
sns.set_style("white")
hist=sns.histplot(data=df_stack, x="Tm", hue='source', bins=15)
hist.set_title('CTRL')
sns.despine()
plt.show()
```
